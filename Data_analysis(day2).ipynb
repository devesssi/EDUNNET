q:find total energy consumption when temp >28 and there was holiday

```python
total_energy_consumption = df[(df["Temperature"] > 28) & (df["Holiday"] == "Yes")]["EnergyConsumption"].sum()
print(f"Total energy consumption when temperature > 28 and it was a holiday: {total_energy_consumption}")
```

Q: create a colmn season (codn: temp< 28--> cold day or temp>28 --> hot day)

```python
def season(x):
  if x>=28:
   return "Hot-day"
  else:
    return "cold-day"
df["Season"]=df["Temperature"].map(season)
print(df["Season"])
```
